```
#!/usr/bin/env python2
# -*- coding: utf-8 -*-
"""
Created on Wed Dec 19 19:48:39 2018

@author: Lyn
"""

import warnings
warnings.filterwarnings("ignore")


import numpy as np
import cmath
import math
import time
from decimal import Decimal


def compute(strike,number,modelname,ref):
    # Fixed Parameters
    S0 = 100
    K = strike
    k = math.log(K)
    r = 0.00
    q = 0.00
    
    # Parameters for FFT 
    
    n = number
    N = 2**n
    
    #step-size
    eta = 0.25
    # damping factor
    alpha = 1.5
    
    # step-size in log strike space
    lda = (2*math.pi/N)/eta;
    
    #Choice of beta
    #beta = np.log(S0)-N*lda/2
    beta = np.log(K)
    #model-specific Parameters
    model = modelname
    params = []     
    if (model == 'GBM'):
        
        sig = 0.25
        params.append(sig);
        
    elif (model == 'VG'):
        
        sig = 0.12
        nu = 0.2
        theta = -0.14
        #
        params.append(sig);
        params.append(nu);
        params.append(theta);
        
    elif (model == 'Heston'):
        
        kappa = 1.5768
        theta = 0.039
        sig = 0.5751
        rho = -0.5711
        v0 = 0.0175
        #
        params.append(kappa)
        params.append(theta)
        params.append(sig)
        params.append(rho)
        params.append(v0)
    print(' ')
    print('===================')
    print('Model is %s' % model)
    print('-------------------')
        
    T = 10.
        
    print(' ')
    start_time = time.time()
    km, cT_km = genericFFT(params, S0, K, r, q, T,alpha, eta, lda, n, model)
    cT_k = cT_km[0]
    
    
    elapsed_time = time.time() - start_time
    print ("K:") + str(K)
    print("N:") + str(N)
    
        
    #cT_k = np.interp(np.log(), km, cT_km)
    print("Option via FFT: for strike %s the option premium is %6.4f" % (np.exp(k), cT_k))
    #print("Option via FFT: for strike %s the option premium is %6.4f" % (np.exp(k), cT_km[0]))
    print('FFT execution time was %0.6f' % elapsed_time)
    error = cT_k - ref
    print("Error%.2E" % Decimal(error))
        
    print(' ')
    start_time = time.time()
    cT_k = evaluateIntegral(params, S0, K, r, q, T, N, k, alpha, eta, n, model)
    elapsed_time = time.time() - start_time
    print("Option via Integration: for strike %s the option premium is %6.4f" % (np.exp(k), cT_k))
    print('Evaluation of integral time was %0.6f' % elapsed_time)
    error = cT_k - ref
    print("Error%.2E" % Decimal(error))
    
    

def generic_CF(u, params, S0, r, q, T, model):
    
    if (model == 'GBM'):
        
        sig = params[0]
        mu = np.log(S0) + (r-q-sig**2/2)*T
        a = sig*np.sqrt(T)
        phi = np.exp(1j*mu*u-(a*u)**2/2)
        
    elif(model == 'Heston'):
        
        kappa  = params[0]
        theta  = params[1]
        sigma  = params[2]
        rho    = params[3]
        v0     = params[4]
        
        tmp = (kappa-1j*rho*sigma*u)
        g = np.sqrt((sigma**2)*(u**2+1j*u)+tmp**2)
        
        pow1 = 2*kappa*theta/(sigma**2)
        
        numer1 = (kappa*theta*T*tmp)/(sigma**2) + 1j*u*T*r + 1j*u*math.log(S0)
        log_denum1 = pow1 * np.log(np.cosh(g*T/2)+(tmp/g)*np.sinh(g*T/2))
        tmp2 = ((u*u+1j*u)*v0)/(g/np.tanh(g*T/2)+tmp)
        log_phi = numer1 - log_denum1 - tmp2
        phi = np.exp(log_phi)
        
        #g = np.sqrt((kappa-1j*rho*sigma*u)**2+(u*u+1j*u)*sigma*sigma)
        #beta = kappa-rho*sigma*1j*u
        #tmp = g*T/2
        
        #temp1 = 1j*(np.log(S0)+(r-q)*T)*u + kappa*theta*T*beta/(sigma*sigma)
        #temp2 = -(u*u+1j*u)*v0/(g/np.tanh(tmp)+beta)
        #temp3 = (2*kappa*theta/(sigma*sigma))*np.log(np.cosh(tmp)+(beta/g)*np.sinh(tmp))
        
        #phi = np.exp(temp1+temp2-temp3);
        

    elif (model == 'VG'):
        
        sigma  = params[0];
        nu     = params[1];
        theta  = params[2];

        if (nu == 0):
            mu = math.log(S0) + (r-q - theta -0.5*sigma**2)*T
            phi  = math.exp(1j*u*mu) * math.exp((1j*theta*u-0.5*sigma**2*u**2)*T)
        else:
            mu  = math.log(S0) + (r-q + math.log(1-theta*nu-0.5*sigma**2*nu)/nu)*T
            phi = np.exp(1j*u*mu)*((1-1j*nu*theta*u+0.5*nu*sigma**2*u**2)**(-T/nu))

    return phi
def evaluateIntegral(params, S0, K, r, q, T, N, k, alpha, eta, n, model):
    
    # Just one strike at a time
    # no need for Fast Fourier Transform
    
    # discount factor
    df = math.exp(-r*T)
    
    sum1 = 0
    for j in range(N):
        nuJ = j*eta
        psi_nuJ = df*generic_CF(nuJ-(alpha+1)*1j, params, S0, r, q, T, model)/((alpha + 1j*nuJ)*(alpha+1+1j*nuJ))
        if j == 0:
            wJ = (eta/2)
        else:
            wJ = eta
        sum1 += np.exp(-1j*nuJ*k)*psi_nuJ*wJ
        
    cT_k = (np.exp(-alpha*k)/math.pi)*sum1
    
    return np.real(cT_k) 
def genericFFT(params, S0, K, r, q, T, alpha, eta, lda, n, model):
    
    N = 2**n
    
    # step-size in log strike space
    gamma = eta*lda/2./np.pi
    
    
    #Choice of beta
    #beta = np.log(S0)-N*lda/2
    beta = np.log(K)
    
    # forming vector x and strikes km for m=1,...,N
    km = np.zeros((N))
    xX = np.zeros((N))
    
    # discount factor
    df = math.exp(-r*T)
    
    nuJ = np.arange(N)*eta
    psi_nuJ = generic_CF(nuJ-(alpha+1)*1j, params, S0, r, q, T, model)/((alpha + 1j*nuJ)*(alpha+1+1j*nuJ))
    
    for j in range(N):  
        km[j] = beta+j*lda
        if j == 0:
            wJ = (eta/2)
        else:
            wJ = eta
        xX[j] = cmath.exp(-1j*beta*nuJ[j])*df*psi_nuJ[j]*wJ
    
    
    y =  np.zeros((2*N),dtype=complex)
    z =  np.zeros((2*N),dtype=complex)
    for i in range(2*N):
        if (i<N):
            y[i] = xX[i]*np.exp(-1j*np.pi*i*i*gamma);
            z[i] = np.exp(1j*np.pi*i*i*gamma);
            
        else:
            z[i] = np.exp(1j*np.pi*((2*N-i-1)**2)*gamma)
   
    
    y_hat = np.fft.fft(y)
    z_hat = np.fft.fft(z)
    
    cs_hat = y_hat*z_hat
    
    
    #frfft 
    
    cs = np.fft.ifft(y_hat*z_hat)
    cs = (np.exp(-1j*np.pi*gamma*i*i) * cs)[:N]
    
    
    #fft 
    yY = np.fft.fft(xX)

    
    cT_km = np.zeros((N))  
    for i in range(N):
        multiplier = math.exp(-alpha*km[i])/math.pi
        #cT_km[i] = multiplier*np.real(cs[i])
        cT_km[i] = multiplier*np.real(yY[i])
    
    
    return km, cT_km

    
K_list = [80., 100.,120.]
N_list = [4,6,8]
model = 'Heston'
ref = [32.5808,22.3189,14.8058]
for i in range(len(K_list)):
    for nu in N_list:
        compute(K_list[i], nu, model, ref[i])
```
